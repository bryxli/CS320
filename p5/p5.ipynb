```python
#q1
import zipfile
import pandas as pd

with zipfile.ZipFile("server_log.zip", "r") as x:
    with x.open("rows.csv", "r") as f:
        df = pd.read_csv(f)
int(sum(df["size"]))
```

```python
#q8
docs = {}
with zipfile.ZipFile("docs.zip", "r") as x:
    for file in x.namelist():
        with x.open(file, "r") as f:
            filing = edgar_utils.Filing(str(f.read()))
            docs[file] = filing
docs["850693/0000850693-07-000159/-index.htm"].dates
```

```python
#q9
states = {}
for filing in docs.keys():
    state = docs[filing].state()
    if state != None:
        if state not in states.keys():
            states[state] = 1
        else:
            states[state] += 1
dict(sorted(states.items(), key=lambda y: y[1], reverse=True))
```

```python
#q12
import re
datetime = pd.to_datetime(df["time"])
each_hour = {}
for i in range(24):
    each_hour[i] = 0
for time in datetime:
    hour = int(re.findall(r"(\d{2}):\d{2}:\d{2}", str(time))[0])
    if hour not in each_hour.keys():
        each_hour[hour] = 1
    else:
        each_hour[hour] += 1
each_hour
```

```python
#q13
import graphviz
country_files = {"Australia":[],"France":[],"Viet Nam":[]}
for ip in access_sortedip.keys():
    region = edgar_utils.lookup_region(ip)
    if region in ["Australia","France","Viet Nam"]:
        ip_df = df[df["ip"] == ip]
        ip_cik = ip_df["cik"].reset_index()
        del ip_cik["index"]
        ip_accession = ip_df["accession"].reset_index()
        del ip_accession["index"]
        ip_extention = ip_df["extention"].reset_index()
        del ip_extention["index"]
        for i in range(len(ip_cik)):
            file = str(int(ip_cik.loc[i]["cik"])) + "/" + str(ip_accession.loc[i]["accession"]) + "/" + str(ip_extention.loc[i]["extention"])
            country_files[region].append(file) 
country_access = {"Australia":set(),"France":set(),"Viet Nam":set()}
for country in country_files.keys():
    files = country_files[country]
    for file in files:
        if file in docs.keys():
            country_access[country].add(docs[file].state())
graph = graphviz.Digraph()
for country in country_access.keys():
    graph.node(country)
    for state in country_access[country]:
        graph.node(state)
        graph.edge(country, state)
graph
```

```python
#q15
from shapely.geometry import Polygon, box, Point
import geopandas as gdp
df = gdp.read_file("shapes/cb_2018_us_state_20m.shp")
locations = gdp.read_file("locations.geojson")
locations_copy = locations.copy().to_crs("epsg:3395")
locations_copy["appearance"] = None
counter = 0
for location in locations_copy["address"]:
    if location in other.keys():
        locations_copy["appearance"][counter] = other[location]
        counter += 1
window = box(-130, 20, -55, 50)
df = df.intersection(window)
df = df[~df.is_empty]
df = gdp.GeoDataFrame(geometry=df)
locations = locations.intersection(window)
locations = locations[~locations.is_empty]
locations = gdp.GeoDataFrame(geometry=locations)
df = df.to_crs("epsg:3395")
locations = locations.to_crs("epsg:3395")
ax = df.plot(color="lightgray")
ax.set_axis_off()
locations["appearance"] = None
for i in range(len(locations_copy["geometry"])):
    if locations_copy["geometry"][i] in locations["geometry"]:
        locations["appearance"][i] = locations_copy["appearance"][i]
locations.plot(ax=ax, cmap="plasma", column="appearance", legend = True)
```
