```python
#q30
import numpy as np
import time
def rand_df(rows):
    return pd.DataFrame(np.random.randint(10, size=(rows, 4)),
                        columns=["A", "B", "C", "D"],
                        index=[f"r{i}" for i in range(1, rows+1)])
def timerows(df):
    t0 = time.time()
    for row in df.iterrows():
        pass
    t1 = time.time()
    return t1 - t0
def timetuples(df):
    t0 = time.time()
    for row in df.itertuples():
        pass
    t1 = time.time()
    return t1 - t0
times_df = pd.DataFrame(dtype=float)
for i in range(1,200):
    df = rand_df(i)
    resultrows = timerows(df)
    resulttuples = timetuples(df)
    times_df.at[i,"iterrows"]=resultrows
    times_df.at[i,"itertuples"]=resulttuples
plot = times_df.plot.line()
plot.set_xlabel("Number of rows")
plot.set_ylabel("Amount of time")
```

```python
#q31
def timeloc(df):
    t0 = time.time()
    total = 0
    for idx in df.index:
        for col in df.columns:
            total += df.loc[idx, col]
    t1 = time.time()
    return t1 - t0
def timeat(df):
    t0 = time.time()
    total = 0
    for idx in df.index:
        for col in df.columns:
            total += df.at[idx, col]
    t1 = time.time()
    return t1 - t0
times_df = pd.DataFrame(dtype=float)
for i in range(1,200):
    df = rand_df(i)
    resultloc = timeloc(df)
    resultat = timeat(df)
    times_df.at[i,"loc"]=resultloc
    times_df.at[i,"at"]=resultat
plot = times_df.plot.line()
plot.set_xlabel("Number of rows")
plot.set_ylabel("Amount of time")
```

```python
#q32
def laugh(x):
    return "ha" * x
def timeloop(df):
    t0 = time.time()
    result = []
    for val in df["A"]:
        result.append(laugh(val))
    t1 = time.time()
    return t1 - t0
def timeapply(df):
    t0 = time.time()
    result = df["A"].apply(laugh).tolist()
    t1 = time.time()
    return t1 - t0
times_df = pd.DataFrame(dtype=float)
for i in range(1,200):
    df = rand_df(i)
    resultloop = timeloop(df)
    resultapply = timeapply(df)
    times_df.at[i,"loop"]=resultloop
    times_df.at[i,"apply"]=resultapply
plot = times_df.plot.line()
plot.set_xlabel("Number of rows")
plot.set_ylabel("Amount of time")
```
